------
Создадим функцию, которая разобьет данные из столбца id и присвоит имена столбцам:

```python
import re

# Функция для разбиения строки с учётом кавычек
def split_row(row):
    return re.split(r',(?=(?:[^"]*"[^"]*")*[^"]*$)', row)

split_data = df_null['id'].apply(split_row).tolist()

new_df = pd.DataFrame(split_data)

column_names = [
    'id', 'client_id', 'order_day', 'category_id', 'category_name', 
    'good_id', 'good_name', 'good_price', 'good_cnt', 'gender_cd', 
    'age', 'education_level', 'city_nm', 'monthly_income_amt', 'steam_popularity_score;'
]
new_df.columns = column_names

new_df
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
# соединим датафреймы
result = pd.concat([df_cleaned, new_df], ignore_index=True) 
result
```

```python
result.info()
```

Вернем некоторым ячейкам их изначальный тип данных:

```python
result['age'] = pd.to_numeric(result['age'], errors='coerce').astype('Int64')
result['category_id'] = pd.to_numeric(result['category_id'], errors='coerce')
result['good_id'] = pd.to_numeric(result['good_id'], errors='coerce')
result['good_price'] = pd.to_numeric(result['good_price'], errors='coerce')
result['good_cnt'] = pd.to_numeric(result['good_cnt'], errors='coerce')
result['steam_popularity_score;'] = pd.to_numeric(result['steam_popularity_score;'], errors='coerce').fillna(0).astype(float)
result.info()
```

```python
cross_tab = pd.crosstab(
    index=result['gender_cd'],
    columns=result['category_name'],
    normalize='index'
).round(4) * 100

# Функция для форматирования таблицы
def format_gender_table(gender_data, gender_name):
    return (gender_data
            .sort_values(ascending=False)
            .head(5)
            .to_frame()
            .rename(columns={gender_name: 'Доля, %'})
            .style
            .format('{:.2f}%')
            .set_caption(f"Топ-5 категорий для '{gender_name}'")
            .background_gradient(cmap='Blues')
           )

# Выводим таблицы по полу
for gender in cross_tab.index:
    display(format_gender_table(cross_tab.loc[gender], gender))
```

Проанализируем разницу между покупателями игровых товаров и остальными клиентами:

```python
df['gamer_status'] = df['category_name'].str.contains('игр', case=False, na=False).map({True: 'Покупатели игр',False: 'Обычные клиенты'})

comparison = df.groupby('gamer_status').agg({
    'age': 'mean',
    'gender_cd': lambda x: (x == 'M').mean(),  # доля мужчин
    'monthly_income_amt': 'mean',
    'good_price': 'mean',
    'order_day': 'nunique' 
}).rename(columns={
    'order_day': 'purchase_frequency',
    'good_price': 'avg_order_price'
})


fig, axes = plt.subplots(2, 2, figsize=(14, 11), dpi=180)

sns.barplot(x=comparison.index, y='age', data=comparison, ax=axes[0, 0])
axes[0, 0].set_title('Средний возраст')

sns.barplot(x=comparison.index, y='gender_cd', data=comparison, ax=axes[0, 1])
axes[0, 1].set_title('Доля мужчин')

sns.barplot(x=comparison.index, y='monthly_income_amt', data=comparison, ax=axes[1, 0])
axes[1, 0].set_title('Доход за месяц')

sns.barplot(x=comparison.index, y='purchase_frequency', data=comparison, ax=axes[1, 1])
axes[1, 1].set_title('Частота покупок');
```

```python
games = df[df['category_name'].str.contains('игр|game', case=False, na=False)]

top_games = (
    games.groupby('good_name')
    .agg({
        'age': 'mean',
        'gender_cd': lambda x: (x == 'M').mean(),  # доля мужчин
        'monthly_income_amt': 'mean',
        'id': 'count'  # количество покупок
    })
    .nlargest(10, 'id')
    .drop(columns='id')
)

heatmap_data = (top_games - top_games.mean()) / top_games.std()  # Нормализация данных

# Форматирование аннотаций
annot_data = pd.DataFrame({
    'возраст': top_games['age'].round(1).astype(str) + ' лет',
    'доля мужчин': (top_games['gender_cd']*100).round(1).astype(str) + '%',
    'доход': (top_games['monthly_income_amt']/1000).round(1).astype(str) + 'k ₽'
})

plt.figure(figsize=(12, 6),dpi=180)
sns.heatmap(
    heatmap_data.T,
    annot=annot_data.T,
    fmt='',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    cbar_kws={'label': 'нормализованные значения (z-score)'}
)
plt.title('Сравнение характеристик аудитории топ-10 игр', fontsize=14)
plt.xlabel('Название игр', fontsize=12)
plt.ylabel('Критерий', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0);
```

# Задача 2
Предположим, что мы задумались над оптимальностью текущих ценовых политик и решили понять, стоит ли улучшить или ухудшить скидки на игры. Ваша задача провести исследование эластичности спроса пользователей к цене, то есть понять, какой будет объем продаж в зависимости от цены. На представленных данных вряд ли можно получить точное предсказание, поэтому необходимо проанализировать свойства спроса по косвенным признакам. 

Ожидаемый результат работы: рекомендации и набор действий по максимизации выручки для издателей видеоигр.

Посмотрим на то, как изменялась средняя цена игр по месяцам и количество продаж:

```python
result['order_day'] = pd.to_datetime(result['order_day'], errors='coerce')

time_trends = result.groupby(result['order_day'].dt.to_period('M')).agg({ #Преобразуем дату в период месяц год
    'good_price': 'mean',
    'good_cnt': 'sum' 
}).reset_index() 
time_trends['order_day'] = time_trends['order_day'].dt.to_timestamp() 

fig, axes = plt.subplots(2, 1, figsize=(12, 8))
sns.lineplot(x='order_day', y='good_price', data=time_trends, ax=axes[0])
axes[0].set_title('Динамика средней цены')
sns.lineplot(x='order_day', y='good_cnt', data=time_trends, ax=axes[1])
axes[1].set_title('Динамика количества продаж')
plt.tight_layout()
```

Вычисляем эластичность:

```python
price_segments = pd.cut(result['good_price'], bins=[0, 100, 500, 1000, np.inf], labels=['<100', '100-500', '500-1000', '>1000'])
result['price_segment'] = price_segments

segment_data = result.groupby('price_segment', observed=False).agg({
    'good_price': 'mean',
    'good_cnt': 'sum'
})

if len(segment_data) > 1:
    elasticity = []
    for i in range(len(segment_data) - 1):
        Q1 = segment_data['good_cnt'].iloc[i]
        Q2 = segment_data['good_cnt'].iloc[i+1]
        P1 = segment_data['good_price'].iloc[i]
        P2 = segment_data['good_price'].iloc[i+1]

        if P1 != 0 and P2 != 0: 
          elasticity.append(((Q2 - Q1) / Q1) / ((P2 - P1) / P1))
        else:
          elasticity.append(np.nan)  # если P1 или P2 равны нулю, то эластичность не определена
    print("Эластичность по ценовым сегментам:", elasticity)
else:
    print("Недостаточно ценовых сегментов для расчета эластичности")
```

---
Посмотрим на топ-5 игр, цены которых нужно повысить/снизить:

```python
# 1. Топ-5 игр для снижения цены 
top_to_discount = result.sort_values(by=['steam_popularity_score;', 'good_cnt'], ascending=[True, True]) \
                        .drop_duplicates(subset=['good_id']) \
                        .head(5)[['good_name', 'good_price', 'good_cnt', 'steam_popularity_score;']]
print("\nТоп-5 игр для снижения цены:")
print(top_to_discount.to_string(index=False))
```

```python
# 2. Топ-5 игр для повышения цены 
top_to_increase = result.sort_values(by=['steam_popularity_score;', 'good_cnt'], ascending=[False, False]) \
                        .drop_duplicates(subset=['good_id']) \
                        .head(5)[['good_name', 'good_price', 'good_cnt', 'steam_popularity_score;']]
print("\nТоп-5 игр для повышения цены:")
print(top_to_increase.to_string(index=False))
```

```python
elasticity_stats = result.groupby('category_name').agg(
    avg_price=('good_price', 'mean'),
    total_sales=('good_cnt', 'sum'),
    popularity_rank=('steam_popularity_score;', 'mean')
).reset_index()
print("\nСводная статистика по эластичности:")
print(elasticity_stats.to_string(index=False))
```
